```python
import matplotlib.pyplot as plt

import sys
from pathlib import Path
import pickle
import pandas as pd
import numpy as np
from itertools import product
from tqdm.notebook import tqdm
from PIL import Image
import os

parent_path = str(Path("../").resolve())
print(parent_path)
sys.path.append(parent_path)

from seads_envs import load_env

%matplotlib inline

os.makedirs("embedded_environments", exist_ok=True)
```

```python
# Render LightsOutReacher
env_name = "LightsOutReacherEnvBs5"
reset_split = "test"
env_kwargs = {
    "max_solution_depth": 2,
    "random_solution_depth": False,
    "max_steps": 50,
    "render_width": 480,
    "render_height": 480}
env = load_env(env_name, reset_split, env_kwargs)
env.seed(30)
env.reset()
rendering = env.render(mode="rgb_array")
plt.imshow(rendering)

img = Image.fromarray(rendering.astype(np.uint8))
img.save('embedded_environments/lightsout_reacher.png')
```

```python
# Render LightsOutJaco
env_name = "LightsOutJacoEnvBs5"
reset_split = "test"
env_kwargs = {
    "max_solution_depth": 2,
    "random_solution_depth": False,
    "max_steps": 50,
    "render_width": 480,
    "render_height": 480,
    "rendering_enabled": True
}
env = load_env(env_name, reset_split, env_kwargs)
env.seed(30)
env.reset()
env.unwrapped._gym_env._camera_id=3
rendering = env.render(mode="rgb_array")
plt.imshow(rendering)
img = Image.fromarray(rendering.astype(np.uint8))
img.save('embedded_environments/lightsout_jaco.png')
```

```python
# Render grid board
from itertools import product

env_name = "LightsOutCursorEnvBs5"
reset_split = "test"
env_kwargs = {
    "max_steps": 10,
    "max_solution_depth": 2,
    "random_solution_depth": False,
}
env = load_env(env_name, reset_split, env_kwargs)
env.seed(30)
env.reset()

fig, ax = plt.subplots(1, 1)
env.plot_board(ax, board_state=env.board_state)
plt.savefig("embedded_environments/lightsout_board.png", bbox_inches = 'tight', pad_inches = 0)
```
